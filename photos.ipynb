```python
import os
import pandas as pd
from tqdm import tnrange, tqdm_notebook
```

```python
factor=100000
filename='Kaggle-renthop.7z'
```

```python
train_df = pd.read_json("train.json")
test_df = pd.read_json("test.json")
tmp=pd.concat([train_df,test_df])
tmp['new_listing_id']=tmp['listing_id']//factor
ids=sorted(tmp['new_listing_id'].unique(),reverse=False)
```

```python
results=list()
for i in tqdm_notebook(ids):
    os.system("7z x ./photos/{} {}*/*.jpg -r".format(filename,i))
    #print("7z x ./photos/{} {}*/*.jpg -r".format(filename,i))
    for j in tqdm_notebook(range(factor),leave=False):
        ii=i*factor+j
        for root, dirs, filenames in os.walk("./images/{}".format(ii)):
            for f in filenames:
                if f.find("jpg")>-1:
                    statinfo = os.stat("./images/{}/{}".format(ii,f))
                    results.append([i,statinfo.st_mtime,statinfo.st_size]) 
            if len(dirs)==1:
                print(dirs[0])
                for root2, dirs2, filenames2 in os.walk("./images/{}/{}".format(dirs[0],ii)):
                    for f in filenames2:
                        if f.find("jpg")>-1:
                            statinfo = os.stat("./images/{}/{}/{}".format(ii,dirs[0],f))
                            results.append([-1*i,statinfo.st_mtime,statinfo.st_size]) 
    os.system("rm -rf images")        
    os.system("7z d ./photos/{} {}*/*.jpg -r".format(filename,i))  
```

```python
outframe=pd.DataFrame(results)
outframe.columns = ['listing_id', 'time','size']
```

```python
outframe.to_csv("./photosmy.csv", sep=',',index=False)
```
